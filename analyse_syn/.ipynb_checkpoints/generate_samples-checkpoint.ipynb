Notebook used to make tests for the experiment reproducing samples.

```python
import numpy as np
import itertools
import os
import pandas as pd
import networkx as nx
import sys
sys.path.append('../')
import tools as tl
import synthetic
```

```python
folders = ['ReciprocityMMSBM_conditionals_20', 'ReciprocityMMSBM_joints_20']

data = []
for f in folders:
    in_folder = f'../../data/input/{f}/'
    nets = os.listdir(in_folder)
    net = [(in_folder, n) for n in nets if n.startswith('syn')]
    data.extend(net)
```

```python
show_details = False
show_plots = False
```

### JointCRep

```python
for d in data:
    '''Import data'''
    in_folder = d[0]
    folder = in_folder.split('/')[4]
    adj = d[1]
    name_id = adj.split('.dat')[0].split('syn')[1]
    
    network = f'../../data/input/{folder}/syn{name_id}.dat'
    A, B, B_T, data_T_vals = tl.import_data(network, ego='source', alter='target', force_dense=False, verbose=False)
    assert len(A)==1
    degrees = [A[0].degree(n) for n in A[0].nodes()]
    k = np.mean(degrees)
    
    '''Import theta inferred'''
    theta = np.load(f'../../data/output_cluster/JointCRep/{folder}/theta{name_id}.npz')
    N, K = theta['u'].shape
    parameters = [theta['u'], theta['v'], theta['w']]
    
    '''Generate five samples with different random seeds'''
    np.random.seed(17)
    rseed = np.random.randint(18)
    out_folder = f'../../data/output_cluster/JointCRep/SAMPLES/{folder}/'

    for i in range(5):
        label = name_id+'_'+str(i)
        synthetic.ReciprocityMMSBM_joints(N=N, K=K, avg_degree=k, show_details=show_details, show_plots=show_plots,
                                          eta=theta['eta'], parameters=parameters, seed=rseed, out_folder=out_folder, 
                                          label=label)
        rseed += np.random.randint(1,18)
        print('\n')
```
