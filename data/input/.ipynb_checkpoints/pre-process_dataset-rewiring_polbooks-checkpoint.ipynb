```python
outnet = 'polbooks'
H = nx.read_gml("polbooks/polbooks.gml")
edge_list = H.edges 
node_list = H.nodes(data=True)
```

### Build multigraph

```python
G = nx.DiGraph()
G.add_edges_from(H.edges()) # or G.add_edges_from(edge_list)
G.add_nodes_from(H.nodes(data=True))
```

```python
list(G.edges(data=True))[:5]
```

```python
list(G.nodes(data=True))[:5]
```

```python
val_list = list(G.nodes('value')) 
```

```python
for ii in val_list: 
    nx.set_node_attributes(G, {ii[0]: ii[1]}, name="com")
```

```python
G.nodes['1000 Years for Revenge']["com"]
```

```python
G.number_of_nodes(),G.number_of_edges()
```

```python
G1 = nx.convert_node_labels_to_integers(G, first_label=0)
```

```python
comm = [G1.nodes[cc]["com"] for cc in list(G1.nodes)]
comm
```

### Remove self loops

```python
G1.remove_edges_from(nx.selfloop_edges(G1))
```

```python
G1.number_of_nodes(),G1.number_of_edges()
```

```python
node_color =[]
for node in G1.nodes(data=True):

    # if the node has the attribute group1
    if 'n' in node[1]['com']:
        node_color.append('blue')

    # if the node has the attribute group1
    elif 'c' in node[1]['com']:
        node_color.append('red')

    # if the node has the attribute group1
    elif 'l' in node[1]['com']:
        node_color.append('green')
```

```python
# umax= [G1.nodes[cc]["com"] for cc in list(G1.nodes) if (G1.nodes[cc]["com"]=='n')]  
```

```python
umax= [G1.nodes[cc]["com"] for cc in list(G1.nodes)] 
nodelist1 = []
nodelist2 = []
N = len(list(G1.nodes))
K = 3
# for k in range(K+1):
for k1 in ['l']:
    nodelist1.extend([i for i in range(N) if umax[i]==k1])
for k2 in ['c']:
    nodelist2.extend([i for i in range(N) if umax[i]==k2])

print('len(nodelist1):', len(nodelist1))
print('len(nodelist2):', len(nodelist2))
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,6))
ax = axes.flatten()

H1 = G1.subgraph(nodelist1)
H2 = G1.subgraph(nodelist2)
pos1 = nx.spring_layout(H1) 
pos2 = nx.spring_layout(H2)  

kk1 = H1
kk2 = H2
nx.draw(kk1,pos=pos1, with_labels=False,node_color = 'blue',cmap = plt.cm.Set1, node_size=25,ax=ax[0])
ax[0].set_axis_off()
nx.draw(kk2,pos=pos2, with_labels=False,node_color = 'red',cmap = plt.cm.Set1, node_size=25,ax=ax[1])
ax[0].set_axis_off()
```
